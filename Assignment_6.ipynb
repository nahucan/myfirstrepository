5. Perform meaningful visualizations using the dataset. Bring at least 3
visualizations.


```python
plt.bar(df['Classification'],df['PL'])
plt.title("Petal Lengths of different Classifications",fontsize=18)
plt.xlabel("Classifications",fontsize=15)
plt.ylabel("Petal Lengths",fontsize=15)
plt.show()
```

```python
sns.lineplot(data=df,x='Classification',y='SL')
plt.title('Sepal Lengths of different Classifications',fontsize=18)
plt.xlabel('Classifications',fontsize=15)
plt.ylabel('Sepal Lengths',fontsize=15)
```

```python
sns.countplot(x='SW',data=df,hue='Classification')
plt.title('Sepal Widths of different Classifications',fontsize=18)
plt.show()
```

```python
df['PW'].value_counts().plot.pie(radius=2)
plt.title('Petal Widths of different Classifications',fontsize=15)
plt.show()
```
